```python
import cv2
import numpy as np
import os
from functools import reduce
from numpy import random
import matplotlib.pyplot as plt
```

```python
def factor(n):
    factors = []
    i = 2
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_list = []
count = 0
n = 2
# while count < 0:
#     if is_prime(n):
#         prime_list.append(n)
#         count += 1
#     n += 1
# print(prime_list)
#print(prime_list)
```
